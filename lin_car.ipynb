# clean data


```python
car['Seats']=car['Seats'].str.split().str.get(0).astype(int)
```

```python
car['engine']=car['engine'].str.split().str.get(0).astype(int)
```

```python
car['ownership'] = car['ownership'].str.extract('(\d+)').astype(int)

```

```python
car['ownership'].unique()
```

```python
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
car['car_prices_in_rupee']=car['car_prices_in_rupee'].str.split().str.get(0)
```

```python

# Convert to numeric (handle errors) and then multiply
car['car_prices_in_rupee'] = pd.to_numeric(car['car_prices_in_rupee'], errors='coerce')
```

```python
# Multiply by 100000 and safely convert to integer
car['car_prices_in_rupee'] = (car['car_prices_in_rupee'] * 100000).round().astype(pd.Int64Dtype())  # Keeps NaN
```

```python
car['transmission'] = car['transmission'].str.replace(r'Automatic|Au', 'Auto', regex=True).str.replace(r'Manual|Ma', 'Man', regex=True)

```

```python
car['car_name']=car['car_name'].str.split().str.slice(0,3).str.join(' ')
```

```python
car.reset_index(drop=True)
```

```python
car.info()
```

```python
mi=car['car_prices_in_rupee'].isna().sum()/len(car)*100
mi     # % of missing value
```

```python
car=car.dropna(subset=['car_prices_in_rupee'])
```

```python
car.drop_duplicates(inplace=True)
```

```python
car.duplicated().sum()
```

```python
car.isnull().sum()
```

```python
car.reset_index(drop=True)
```

```python
car.describe()
```

```python

# Assuming 'car' tera original dataframe hai
# car_name ko brand aur model mein split karo
car['brand'] = car['car_name'].str.split().str[0]  # Pehla word brand
car['model'] = car['car_name'].str.split().str[1:].str.join(' ')  # Baaki words model

# Original car_name drop karo (optional, agar chahiye to rakh sakta hai)
car = car.drop('car_name', axis=1)
```

```python
car.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score,mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X=car.drop(columns=['car_prices_in_rupee'])
y=car['car_prices_in_rupee']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
oh=OneHotEncoder()
oh.fit(X[['fuel_type','transmission','brand','model']])
```

```python
X_train.head()
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=oh.categories_),['fuel_type','transmission','brand','model']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train )
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```

```python
score=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(score))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)

y_pred=pipe.predict(X_test)

r2_score(y_test,y_pred)
```
